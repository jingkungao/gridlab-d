```python
%matplotlib inline

import numpy as np
import pandas as pd

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)

def plotly_plot(x,y,name='Data',title=''):
    if len(x)>20000:
        print('Too many points, please reduce number of points! (less than 20000)')
        return
    trace1 = go.Scatter(
        name = name,
        x = x,
        y = y)

    fig = go.Figure(
        data = [trace1],
        layout = go.Layout(showlegend=True,
                          title=title))

    iplot(fig)
```

```python
raw_freq = pd.read_csv('frequency.PLAYER',index_col=0,names=['time','freq[Hz]'],
                       parse_dates=True,infer_datetime_format=True)
```

```python
freq_low = 59.96
freq_high = 60.04

ax = raw_freq.plot(figsize=(12,4),rot=30)
ax.axhline(y=freq_low, c='red')
ax.axhline(y=freq_high, c='green')
```
